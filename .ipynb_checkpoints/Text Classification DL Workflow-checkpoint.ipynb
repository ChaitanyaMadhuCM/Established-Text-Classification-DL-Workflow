# 1. Hypothesis Generation 

- **Problem Statement: Build a DL Model (from scratch) that takes News Text as Input and Categorizes it into one of the 20 pre-defined news categories.**

- **Use-Case:** Multi-Class Text Classification Problem

- **Input Variables:**
    - 1. News Text (Text Field with average length of 720 characters) 
- **Output Variables:**
    - 1. News Category (One of the 20 pre-defined Categories)


# 2. Import Libraries 

```python
# Data handling libraries
import pandas as pd
import numpy as np

# data visualization libraries 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# neural network libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GlobalAveragePooling1D, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping
from keras.models import Sequential
from keras.layers import Embedding, GlobalAveragePooling1D, Dense, LSTM, SpatialDropout1D
from keras.utils import plot_model
from tensorflow.keras.models import load_model

import nltk
nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# model dumping libaries 
import joblib

import re
import time
import warnings
warnings.filterwarnings('always')
warnings.filterwarnings('ignore')
```

# 3. Import Dataset 

```python
# Load the 20 Newsgroups dataset
newsgroups_data = fetch_20newsgroups( subset='all', remove=('headers', 'footers', 'quotes'))
X = newsgroups_data.data
y = newsgroups_data.target
```

```python
# Access label names
label_names = newsgroups_data.target_names

# Create a mapping between label names and numerical values
label_mapping = {label_name: idx for idx, label_name in enumerate(label_names)}

# Display the mapping
print("Label Mapping:")
for label_name, numerical_value in label_mapping.items():
    print(f"{label_name}: {numerical_value}")
```

```python
print(len(X))
print(len(y))

# view the first sample from the dataset
print(X[0])
print(y[0])
```

# 4. Exploratory Data Analysis 

```python
# Print List of Unique Values
unique_values = np.unique(y)
print('a) List of Unique Values:', unique_values)

# Print Count of Unique Values
unique_values, counts = np.unique(y, return_counts=True)
print('b) Count of Unique Values:')
print(np.column_stack((unique_values, counts)))

# Print Normalized Count of Unique Values
normalized_counts = counts / counts.sum()
print('c) Normalized Count of Unique Values:')
print(np.column_stack((unique_values, normalized_counts)))
```

# 5. Data Preprocessing 

```python
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove special characters and digits
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenization and remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = nltk.word_tokenize(text)
    tokens = [token for token in tokens if token not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Join tokens back into text
    preprocessed_text = ' '.join(tokens)

    return preprocessed_text

# Apply preprocessing to each document in X
X_preprocessed = [preprocess_text(doc) for doc in X]

# Print original and preprocessed text for the first document
print("Original Text:")
print(X[0])

print("\nPreprocessed Text:")
print(X_preprocessed[0])
```

# 6. Dataset Splitting 

```python
# Split the data into training and validation sets 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
```

# 7. Data Transformation (for NLP) 

```python
# Convert NumPy array elements to strings
X_train_str = np.vectorize(str)(X_train)
X_test_str = np.vectorize(str)(X_test)

# Tokenize and pad the sequences
max_words = 10000
maxlen = 100

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X_train_str)

X_train_seq = tokenizer.texts_to_sequences(X_train_str)
X_test_seq = tokenizer.texts_to_sequences(X_test_str)
X_train_pad = pad_sequences(X_train_seq, maxlen=maxlen)
X_test_pad = pad_sequences(X_test_seq, maxlen=maxlen)
```

```python
# y is already in encoded form when loaded, and hence no explicit encoding is required. 
```

# 8. Model Selection and Fine-Tuning 

```python
### Define Network ###
model = Sequential()

model.add(Embedding(input_dim=max_words, output_dim=64, input_length=maxlen))
model.add(GlobalAveragePooling1D())
model.add(Dense(128, activation='relu'))
model.add(Dense(20, activation='softmax'))  # 20 output classes for 20 Newsgroups

print(model.summary())  # visualization for simple models
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)  # visualization for complex models


### Compile Network ###
model.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])


# Fit Network - Training ###
start_time = time.time()
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
history = model.fit(X_train_pad, y_train, epochs=10, batch_size=64, validation_split=0.2, callbacks=[early_stopping])
end_time = time.time()
```

```python
### Evaluate Network ### 

test_loss, test_acc = model.evaluate(X_test_pad, y_test)
print(f"Test Accuracy: %.4f" % (test_acc*100))
print(f"Training time: {end_time - start_time} seconds")
```

```python
# Model Visualization - Plot learning curves with the history object from model.fit 

print(history.history.keys())

# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

# 9. Model Dumping 

```python
# Save the trained model and related files 
dump_path = f"model_dump/"

# Save the entire model to a file
model.save(f"{dump_path}/dl_model.h5")
# Save the tokenizer to a file using joblib
joblib.dump(tokenizer, f"{dump_path}/tokenizer.joblib")
# Save the class_labels dictionary using joblib
joblib.dump(label_mapping, f"{dump_path}/class_labels.joblib")

print("Model and related files are saved successfully !")
```

```python
# load the saved model and supporting files
dump_path = f"model_dump/"
loaded_model = load_model(f"{dump_path}/dl_model.h5")
loaded_tokenizer = joblib.load(f"{dump_path}/tokenizer.joblib")
loaded_class_labels = joblib.load(f"{dump_path}/class_labels.joblib")

maxlen = 100 # value from the training phase 

# Use the loaded tokenizer to tokenize new data
new_data = ["""\nReflect!\n\nIt's the GPU processing the parallel computations NOT the memory...\n\nThe GPU can handle parallel computations involving data 
from any of its attached devices when it needs to.\n\nAn important feature of GPUs is the ability to parallelize tasks. This allows the GPU to 
execute multiple tasks concurrently, freeing up resources for other computations. This is particularly useful in a multi-tasking environment 
where several parallel tasks need to be executed simultaneously. While each task is processing data, the GPU resources are available for other 
computations. Once tasks are ready to transfer their results, they can access the GPU and perform the computations.\n\nIn contrast, on a 
single-core CPU, when a computation is initiated, the core is occupied until the task is completed. This leads to a sequential lockout for 
other processes, regardless of their computational requirements."""]

# Apply pre-processing on new data using the preprocess_text function defined in the training phase 
new_data_preprocessed = [preprocess_text(doc) for doc in new_data]

new_data_sequences = loaded_tokenizer.texts_to_sequences(new_data_preprocessed)
new_data_padded = pad_sequences(new_data_sequences, maxlen=maxlen)

# Perform inference using the loaded model
predicted_probabilities = loaded_model.predict(new_data_padded)

# Get the label names and numbers
label_names = list(loaded_class_labels.keys())
label_numbers = list(loaded_class_labels.values())

# Display the predicted class details and detailed explanation for predicted probabilities
for i in range(len(new_data_preprocessed)):
    print(f"\nSample {i + 1}:")
    
    # Get the predicted class details
    predicted_class_index = np.argmax(predicted_probabilities[i])
    predicted_class = label_numbers[predicted_class_index]
    predicted_class_label = label_names[predicted_class_index]
    confidence_percentage = predicted_probabilities[i, predicted_class_index] * 100
    
    # Display the predicted class details
    print(f"Predicted Class: {predicted_class}")
    print(f"Predicted Class Label: {predicted_class_label}")
    print(f"Prediction Confidence: {confidence_percentage:.2f}%")
    
    # Detailed explanation of prediction probabilities for all classes
    print("\nDetailed Explanation of Predicted Probabilities:")
    for label_name, label_number, probability in zip(label_names, label_numbers, predicted_probabilities[i]):
        print(f"{label_name} ({label_number}): {probability:.4f}")
    print("\n\n")
    
```
